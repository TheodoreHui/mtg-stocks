```python
cols = ["name", "mana_cost", "cmc", "colors", "color_identity","legalities", "type", "supertypes", 
        "subtypes", "rarity", "text", "set", "set_name", "id"]
def setCardDict(card, cols):
    card_dict = {}
    for col in cols:
        try:
            card_dict[col] = getattr(card, col)
        except:
            card_dict[col] = None
            
    return card_dict
```

```python
all_cards = Card.all()
all_cards
```

```python
from tqdm import tqdm
all_cards_list  = []
names = set()
# setup card dataframe with all columns
all_cards_df = pd.DataFrame(columns=cols)
for card in tqdm(all_cards):
    card_dict = setCardDict(card, cols)
    if not (card_dict['name'] in names):
        try:
            #if card_dict['legalities'][0]['legality'] == 'Legal':
            names.add(card_dict['name'])
            all_cards_list.append(card_dict)
        except TypeError as e:
            continue
    
len(all_cards_list)
```

```python
all_cards_df = pd.DataFrame(all_cards_list)
all_cards_df.to_csv("../data/all_mtg_cards.csv", index=False)

print("File output!")
```
